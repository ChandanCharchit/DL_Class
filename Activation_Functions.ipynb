```
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  
    return s,ds

x=np.arange(-6,6,0.01)
y=np.arange(-0.5,1.5,0.01)
sigmoid(x)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,np.zeros(x.shape),color = 'pink')
ax.plot(np.zeros(y.shape),y,color = 'pink')
ax.plot(x,sigmoid(x)[0], color="#f6815a", linewidth=3, label="sigmoid")
ax.plot(x,sigmoid(x)[1], color="#58dfb3", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
fig.show()
```

```
def relu(z):
  t1 = (np.maximum(z,0))
  t2 = (np.maximum(z,0.1))
  return t1,t2

z=np.arange(-4,4,0.01)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,np.zeros(x.shape),color = 'pink')
ax.plot(np.zeros(y.shape),y,color = 'pink')
ax.plot(z,relu(z)[0], color="#ee5c6b", linewidth=3, label="ReLu")
ax.plot(z,relu(z)[1], color="#a2e356", linewidth=3, label="Leaky Relu")
ax.legend(loc="upper right", frameon=False)
fig.show()
```

```
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt

z=np.arange(-4,4,0.01)
y=np.arange(-2,2,0.01)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,np.zeros(x.shape),color = 'pink')
ax.plot(np.zeros(y.shape),y,color = 'pink')
ax.plot(z,tanh(z)[0], color="#307EC7", linewidth=3, label="tanh")
ax.plot(z,tanh(z)[1], color="#a2e356", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
fig.show()
```
